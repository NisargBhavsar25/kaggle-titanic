```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
df.info()
```

```python
ndf = df.dropna(subset=['Embarked'])

ndf.info()
```

```python
ndf["Age"].fillna(ndf["Age"].mean(), inplace = True)

ndf.info()
```

```python
ndf["Cabin"].fillna("N", inplace = True)

ndf.info()
```

```python
for x in ndf.index:
    ndf["Cabin"][x] = ndf["Cabin"][x][0]
```

```python
print(ndf["Cabin"])
```

```python
for x in ndf.index:
    if ndf["Cabin"][x] == "A":
        ndf["Cabin"][x] = -0.845
    elif ndf["Cabin"][x] == "B":
        ndf["Cabin"][x] = -0.595
    elif ndf["Cabin"][x] == "C":
        ndf["Cabin"][x] = -0.375
    elif ndf["Cabin"][x] == "D":
        ndf["Cabin"][x] = -0.125
    elif ndf["Cabin"][x] == "E":
        ndf["Cabin"][x] = 0.125
    elif ndf["Cabin"][x] == "F":
        ndf["Cabin"][x] = 0.375
    elif ndf["Cabin"][x] == "G":
        ndf["Cabin"][x] = 0.595
    elif ndf["Cabin"][x] == "N":
        ndf["Cabin"][x] =0.845
```

```python
nndf = ndf.drop(columns=['Cabin'])
```

```python
nndf.head()
```

```python
nndf["Age"] = (nndf["Age"] - nndf["Age"].min()) / (nndf["Age"].max() - nndf["Age"].min())
nndf["SibSp"] = (nndf["SibSp"] - nndf["SibSp"].min()) / (nndf["SibSp"].max() - nndf["SibSp"].min())
nndf["Parch"] = (nndf["Parch"] - nndf["Parch"].min()) / (nndf["Parch"].max() - nndf["Parch"].min())
nndf["Fare"] = (nndf["Fare"] - nndf["Fare"].min()) / (nndf["Fare"].max() - nndf["Fare"].min())
```

```python
nndf.head()
```

```python
nndf["Pclass"] = nndf["Pclass"].astype('float')
nndf.info()
```

```python
for x in nndf.index:
    if nndf["Pclass"][x] == 1:
        nndf["Pclass"][x] = -1
    elif nndf["Pclass"][x] == 2:
        nndf["Pclass"][x] = 0
    elif nndf["Pclass"][x] == 3:
        nndf["Pclass"][x] = 1
```

```python
nndf.head()
```

```python
for x in nndf.index:
    if nndf["Embarked"][x] == "C":
        nndf["Embarked"][x] = -1
    elif nndf["Embarked"][x] == "Q":
        nndf["Embarked"][x] = 0
    elif nndf["Embarked"][x] == "S":
        nndf["Embarked"][x] = 1
```

```python
nndf.head()
```

```python
nndf.drop(columns=['Name'], inplace=True)
nndf.head()
```

```python
nndf.drop(columns=['Ticket'], inplace=True)
nndf.head()
```

```python
for x in nndf.index:
    if nndf["Sex"][x] == "male":
        nndf["Sex"][x] = 0
    elif nndf["Sex"][x] == "female":
        nndf["Sex"][x] = 1
```

```python
nndf.head()
```

```python
nndf.set_index('PassengerId')
print(nndf.info())
```

```python
nndf.to_csv('ready_train.csv')
```
