```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('test.csv')
df.head()
```

```python
df.info()
```

```python
df["Age"].fillna(df["Age"].mean(), inplace = True)

df.info()
```

```python
df["Cabin"].fillna("N", inplace = True)

df.info()
```

```python
for x in df.index:
    df["Cabin"][x] = df["Cabin"][x][0]
```

```python
print(df["Cabin"])
```

```python
for x in df.index:
    if df["Cabin"][x] == "A":
        df["Cabin"][x] = -0.845
    elif df["Cabin"][x] == "B":
        df["Cabin"][x] = -0.595
    elif df["Cabin"][x] == "C":
        df["Cabin"][x] = -0.375
    elif df["Cabin"][x] == "D":
        df["Cabin"][x] = -0.125
    elif df["Cabin"][x] == "E":
        df["Cabin"][x] = 0.125
    elif df["Cabin"][x] == "F":
        df["Cabin"][x] = 0.375
    elif df["Cabin"][x] == "G":
        df["Cabin"][x] = 0.595
    elif df["Cabin"][x] == "N":
        df["Cabin"][x] =0.845
```

```python
ndf = df.drop(columns=['Cabin'])
```

```python
ndf.head()
```

```python
ndf["Age"] = (ndf["Age"] - ndf["Age"].min()) / (ndf["Age"].max() - ndf["Age"].min())
ndf["SibSp"] = (ndf["SibSp"] - ndf["SibSp"].min()) / (ndf["SibSp"].max() - ndf["SibSp"].min())
ndf["Parch"] = (ndf["Parch"] - ndf["Parch"].min()) / (ndf["Parch"].max() - ndf["Parch"].min())
ndf["Fare"] = (ndf["Fare"] - ndf["Fare"].min()) / (ndf["Fare"].max() - ndf["Fare"].min())
```

```python
ndf.head()
```

```python
ndf["Pclass"] = ndf["Pclass"].astype('float')
ndf.info()
```

```python
ndf["Fare"].fillna(df["Fare"].mean(), inplace = True)

ndf.info()
```

```python
for x in ndf.index:
    if ndf["Pclass"][x] == 1:
        ndf["Pclass"][x] = -1
    elif ndf["Pclass"][x] == 2:
        ndf["Pclass"][x] = 0
    elif ndf["Pclass"][x] == 3:
        ndf["Pclass"][x] = 1
```

```python
ndf.head()
```

```python
for x in ndf.index:
    if ndf["Embarked"][x] == "C":
        ndf["Embarked"][x] = -1
    elif ndf["Embarked"][x] == "Q":
        ndf["Embarked"][x] = 0
    elif ndf["Embarked"][x] == "S":
        ndf["Embarked"][x] = 1
```

```python
ndf.head()
```

```python
ndf.drop(columns=['Name'], inplace=True)
ndf.head()
```

```python
ndf.drop(columns=['Ticket'], inplace=True)
ndf.head()
```

```python
for x in ndf.index:
    if ndf["Sex"][x] == "male":
        ndf["Sex"][x] = 0
    elif ndf["Sex"][x] == "female":
        ndf["Sex"][x] = 1
```

```python
ndf.head()
```

```python
ndf.set_index('PassengerId')
print(ndf.info())
```

```python
ndf.to_csv('ready_test.csv')
```
